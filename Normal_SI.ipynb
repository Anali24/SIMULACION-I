Sus características son las siguientes:



•       Es una distribución simétrica.

•       Es asintótica, es decir sus extremos nunca tocan el eje horizontal, cuyos valores tienden a infinito.

•       En el centro de la curva se encuentran la media, la mediana y la moda.

•       El área total bajo la curva representa el 100% de los casos.

•       Los elementos centrales del modelo son la media y la varianza.



Esta distribución es un modelo matemático que permite determinar probabilidades de ocurrencia para distintos valores de la variable.


```python
import matplotlib.pyplot as plt
import random as rnd
import numpy as np
```

```python
#Distrubución Normal
def f(a, b, n):
  lista = []  # Creamos una lista vacía para almacenar los valores de la muestra
  # Generamos n números aleatorios y aplicamos una transformación para obtener una muestra de la distribución
  for i in range(n):
    U = rnd.random()  # Generamos un número aleatorio U entre 0 y 1
    # Aplicamos la transformación para obtener valores de la distribución normal
    X = (U ** 0.135 - (1 - U) ** 0.135) / 0.1975
    lista.append(X)  # Añadimos el valor X a la lista
  return lista  # Devolvemos la lista con la muestra de la distribución normal
```

```python
# Generamos una muestra de distribución normal con parámetros a=10, b=15 y n=100000
G = f(10, 15, 100000)

# Imprimimos la muestra generada
#print(G)
```

```python
# Calculamos la media de la muestra G
media = np.mean(G)
print(media)
```

```python
# Calculamos la varianza de la muestra G
varianza = np.var(G)

# Imprimimos la varianza calculada
print(varianza)
```

```python
# Crear un histograma de la muestra con densidad (normalizado)
plt.hist(G, bins=100, density=True, alpha=0.6, color='g', label='Histograma de la muestra')

# Generar puntos x para la función de densidad
x_vals = np.linspace(min(G), max(G), 1000)

# Calcular la función de densidad de probabilidad (FDP) teórica de la muestra generada
def normal_pdf(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Calcular los valores de la función de densidad para los puntos x
y_vals = normal_pdf(x_vals, media, np.sqrt(varianza))

# Graficar la función de densidad teórica
plt.plot(x_vals, y_vals, 'r', label='Función de densidad teórica')

# Añadir etiquetas y título
plt.xlabel('Valores de la muestra')
plt.ylabel('Densidad de probabilidad')
plt.title('Función de densidad de probabilidad de una distribución normal')
plt.legend()

# Mostrar el gráfico
plt.show()
```

Gráfica :La gráfica de su función de densidad tiene una forma acampanada y es simétrica respecto de un determinado parámetro estadístico. Esta curva se conoce como campana de Gauss y es el gráfico de una función gaussiana.

```python
#Obtenemos distribución Acumulativa
#Obtenemos distribución Acumulativa
def cdf(G):
  n = len(G)           # Calculamos la cantidad de elementos en G
  x = np.sort(G)       # Ordenamos los elementos de G de menor a mayor
  y = np.arange(0, n) / n  # Creamos un array con valores desde 0 hasta n-1 divididos por n
  return x, y           # Devolvemos los valores ordenados x y los valores de la CDF y
```

```python
# Asignamos los valores retornados por cdf(G) a las variables X y Y
X,Y =cdf(G)
```

```python
# Graficamos la función de distribución acumulada (CDF) usando un gráfico de paso (step plot)
plt.step(X, Y)

# Añadimos una línea vertical en la posición de la media, con color verde
plt.axvline(media, color="green")

# Añadimos una línea horizontal en y=0.5 (CDF=0.5), con color rojo
plt.axhline(0.5, color="red")

#Etiquetas y títulos
plt.title('Función Acumulativa')
# Mostramos el gráfico completo
plt.show()
```
