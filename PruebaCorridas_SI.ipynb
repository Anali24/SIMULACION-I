# Prueba de corridas
Prueba de Corridas:

Prueba para verificar la aleatoriedad de un conjunto de números, en este caso, se realizó con la siguiente hipotesis:

Ho := Los numeros de la muestra son independientes.
H1 := No Ho

Procedimiento:

Paso 1: Determinar una secuencia S de 1´s y 0's . Es 0 si Xi<= Xi-1, en caso contrario se escribe 1.

Paso 2:

Halle el número de corrisas (rachas observadas)

Paso 3:

Calcule el valor esperado y la varianza del número de corridas, así como el estadístico Z0

Valor esperado = (2n-1)/3

Varianza= (16-29)/90

Zo= Co- valor esperado/ varianza

Si Zo vive se acpeta Ho.

Se importan las librerias a usar, numpy para arreglos, prettytable para crear tablas y math para realizar operaciones.

```python
import numpy as np
from prettytable import PrettyTable
import math

datos = [
    0.048, 0.591, 0.478, 0.166, 0.932,
    0.926, 0.007, 0.551, 0.362, 0.533,
    0.368, 0.437, 0.178, 0.411, 0.545,
    0.787, 0.594, 0.086, 0.951, 0.772,
    0.298, 0.199, 0.157, 0.401, 0.631
]

# Determinamos una secuencia de unos y ceros.
def secuencia(datos):
    S = []
    for i in range(len(datos) - 1):  # Iterar hasta el penúltimo elemento
        if datos[i+1] <= datos[i]:
            S.append(0)  # Ingresamos en la lista S, los 0's observados
        else:
            S.append(1)  # Ingresamos en la lista S, los 1's observados
    return S

S = secuencia(datos)

tabla = PrettyTable()
tabla.field_names = ["Xi", "S"]

for i in range(len(S)):
    tabla.add_row([i+1, S[i]])

print(tabla)

# Ahora hay que contar los 0's y 1's en la lista, recordando que:
# Si tenemos 0's o 1's consecutivos, cuenta como 1 corrida.

def corridas(secuencia):
    # Inicializar variables para contar las corridas
    contador_corrida_ceros = 0
    contador_corrida_unos = 0

    # Recorrer la lista y contar las corridas
    for i in range(len(secuencia)):
        # Contar corridas de ceros
        if secuencia[i] == 0:
            if i == 0 or secuencia[i - 1] != 0:
                contador_corrida_ceros += 1
        # Contar corridas de unos
        elif secuencia[i] == 1:
            if i == 0 or secuencia[i - 1] != 1:
                contador_corrida_unos += 1
    return contador_corrida_ceros, contador_corrida_unos

contador_corrida_ceros, contador_corrida_unos = corridas(S)

print("Cantidad de corridas de ceros:", contador_corrida_ceros)
print("Cantidad de corridas de unos:", contador_corrida_unos)

total_corridas = contador_corrida_ceros + contador_corrida_unos
print("Cantidad total de corridas:", total_corridas)

n = len(datos)

# Calculamos el valor esperado
valor_esperado = ((2 * n) - 1) / 3

# Calculamos la varianza
varianza = ((16 * n) - 29) / 90

# Calculamos Zo
Zo = (total_corridas - valor_esperado) / math.sqrt(varianza)

# Imprimimos los resultados
print("Valor esperado:", valor_esperado)
print("Varianza:", varianza)
print("Zo:", Zo)
Z = 1.96
print("Z:", Z)
#Hacemos la acomparación, para ver si se acepta la hipótesis
if -1*Z <= Zo and Zo <= Z:
  print("Por lo tanto, se acepta la hipotesis.")
else:
  print("Por lo tanto, se rechaza la hipoteis.")
```
