```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(xcero,gama,n):
  lista =[]
  #genero una lista vacia para ver que genera la muestra for i in range(n):
  for i in range (n):
    U=rnd.random()
    #X es el mapeo (2. Devolver x)
    X=gama*np.tan(np.pi*(U-1/2))+xcero
    lista.append(X)
  return lista
```

```python

G = f(-2,1.0,1000)
print(G)
```

```python
media = np.mean(G)
print(media)
```

```python
varianza =np.var(G)
print(varianza)
```

```python
plt.hist(G,bins=100)
#densidad
```

```python
plt.plot(G)
```

```python
def cdf(G):
  n=len(G)
  x=np.sort(G)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y =cdf(G)
```

```python
plt.step(X,Y)
#plt.axvline(media, color="green")
plt.axhline(0.5, color="red")
```
