```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimacion():
    u1 = np.random.uniform(0, 1)
    u2 = np.random.uniform(0, 1)
    if u2 < (1 - u1**2)**0.5:
        return 1  # éxito
    else:
        return 0  # fracaso
```

```python
# Parámetros
N = 20  # Número de veces que hacemos el procedimiento

# Listas para guardar los resultados
p = []  # Lista vacía para guardar los elementos y obtener probabilidad
suma = 0  # Contador para la suma de éxitos

# Bucle para realizar las estimaciones
for i in range(N):
    resultado = estimacion()
    suma += resultado
    p.append(suma / (i + 1))  # i + 1 porque en Python iniciamos en 0

# Graficar la probabilidad
plt.plot(p)
plt.title('Probabilidades - Estimacion')
plt.xlabel('Número de iteraciones')
plt.ylabel('Probabilidad')
plt.show()
```

```python
#Cálculo de Pi
pi=4*suma/N
print(pi)
```
