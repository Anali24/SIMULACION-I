***Estmación del Número Pi.***

Solución Analítica:

image.png




Sabemos que la ecuación de un circulo está dada por:

$x^2 + y^2 = 1 $

Así entonces despejamos  y.

$ \implies  y= \sqrt{1-x^2}$

Elegimos a el radio r=1, para facilitar los calculos.

Sabemos que la estimación de pi,está dado por:

$\frac{A_{circulo}}{A_{cuadrado}} \approx \frac{n}{N}$

Con N el Número total de simulaciones y n el número de éxitos.



image.png

Así entonces :

$
A_{cuadrado} = 1
$
$
A_{circulo} = 1/4
$

$ \implies \frac{n}{N} ≈ \frac{\pi}{4} $
$\implies \pi ≈ \frac{4n}{N} $



***Analíticamente:***

Consideramos la variables:

La función $x $ puede definirse como:
$ X =
    \begin{cases}
      1 & \text{si }  \text{ cae dentro del círculo}, (p) éxito \\
      0 & \text{si }  \text{ cae fuera del círculo}.  (1 - p ) fracaso
    \end{cases}
$

Ahora para:

$ z = 4 \frac{x_1 + x_2 + x_3 + .. + x_N}{N} $

Calculamos el valor esperado y varianza:

E[z] = $\frac{4}{N}( E[X_1] + .... + E[X_N]) $

$ \implies E_[z] = 4p $  

Para cualquier momento:

$E[x_i] = 1*p + 0* (1-p) =p$


$ V[x_i] = E[x_i ^2] - E[x_i] ^2 $

$ = p - p^2 = p(1-p) $

Estimador insezgado :Da ek valor que esperamos:

$V[z] = \frac{16}{N^2}(V[x_1] +V[x_2]+..V[x_N]   ) $

$\implies \frac{16}{N^2} Np(1-p)  $

$ \therefore V[z] = \frac{16N(1-p)}{N} $

*Usando la Desigualdad de Chebyshev Obtenemos:*

$P( | z_N - \pi | \geq \epsilon ) \leq \frac{16p ( 1-p)}{N\epsilon ^2} $



***Simulación:***

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimacion():
    u1 = np.random.uniform(0, 1)
    u2 = np.random.uniform(0, 1)
    if u2 < np.sqrt(1 - u1**2):
        return 1  # éxito
    else:
        return 0  # fracaso
```

```python
# Parámetros
N = 1000  # Número de veces que hacemos el procedimiento (más grande para mayor precisión)

# Listas para guardar los resultados
p = []  # Lista vacía para guardar los elementos y obtener probabilidad
suma = 0  # Contador para la suma de éxitos
```

```python
# Bucle para realizar las estimaciones
for i in range(N):
    resultado = estimacion()
    suma += resultado
    p.append(4 * suma / (i + 1))  # Estimación de Pi
```

```python
# Cálculo final de Pi
pi_estimado = 4 * suma / N
print(f"Estimación final de π : {pi_estimado}")
```

```python
# Graficar la probabilidad
plt.plot(p)
plt.title('Estimación de π')
plt.xlabel('Número de iteraciones')
plt.ylabel('Estimación de π')
plt.axhline(y=np.pi, color='r', linestyle='--')  # Línea horizontal en π
plt.show()
```

Notemos que se ve una muestra de tamaño N$\approx 500$, nos da un resultado aceptable.

Histograma:


```python
# Graficar del histograma
plt.hist(p, bins = 50 , density= True)
plt.axvline(np.pi , color ='r')
plt.title('Estimación de π')
plt.show()
```

Calculamos la media y la varianza(para obtener la desviación estandar de la estimación:



```python
media = np.mean(p)
desviacion = np.std(p)

print(f"Media : {media}")
print(f"Desviación Estándar : {desviacion}")
```

Notemos que la media se aproxima bien y la desviación es nuestro error
