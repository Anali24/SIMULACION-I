Laplace Inversa

La distribución de Laplace resulta de la diferencia de dos variables exponenciales aleatorias, independientes e idénticamente distribuidas.

```python
import matplotlib.pyplot as plt
import random as rnd
import numpy as np
import math as mt
```

```python

def f(a,b,n) :
  lista =[]
  #genero una lista vacia para ver que genera la muestra for i in range(n):
  for i in range (n):
    U=rnd.random()
    #X es el mapeo (2. Devolver x)
    X=mt.tan((mt.pi)*U)
    lista.append(X)
  return lista
```

```python
G = f(10,15,100000)
print(G)
```

```python
media = np.mean(G)
print(media)
```

```python
varianza =np.var(G)
print(varianza)
```

Definimos una función donde:

 CDF calcula la función de distribución acumulativa y con sort ordenamos los elementos de G en orden ascendente utilizando la función

```python
def cdf(G):
  n=len(G)
  x=np.sort(G)
  y=np.arange(0,n)/n
  return x,y
```

```python
X,Y =cdf(G)
```

Gráfica de Distribución

```python
plt.step(X,Y)
plt.axvline(media, color="green")
plt.axhline(0.5, color="red")
```
