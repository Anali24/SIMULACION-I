```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parámetros
a=4
b=32
c=40
```

Función de Densidad


Esta se obtiene mediante la gráfica, con ayuda de la recta tangente

```python
def f(x,a,b,c):
  if a<=x<b:
    return (2*(x-a)/((c-a)*(b-a)))
  elif b<=x<=c:
    return (2*(x-c)/((c-a)*(b-c)))
  else:
    return 0
```

```python
x_vals = np.linspace(a, c, 1000000)
y_vals = [f(x,a,b,c) for x in x_vals]
```

```python
plt.plot(x_vals, y_vals, color='blue')
plt.show()
```

Funcion Acumulada

Se hace con la función de Densidad, y la derivamos

```python
def F(x,a,b,c):
  if x<a:
    return 0
  elif a<=x<b:
    return ((x-a)**2/((c-a)*(b-a)))
  elif b<=x<=c:
    return 1-(c-x)**2/((c-a)*(c-b))
  elif x>c:
    return 1
```

```python
x1_vals = np.linspace(a, c, 1000000)
y1_vals = [F(x,a,b,c) for x in x1_vals]
```

```python
plt.plot(x1_vals, y1_vals, color='blue')
plt.show()
```

Función Inversa

Para la función inversa, igualamos a R y despejamos a x, al final R--->u

```python
def F_i(x,a,b,c):

  if 0<=x<(c-a)/(b-a):
    return a+(x*(c-a)*(b-a))**(1/2)
  elif (c-a)/(b-a)<=x<=1:
    return c-((c-a)*(c-b)*(1-x))**(1/2)
```

```python
y2_vals = np.linspace(0, 1, 1000000)
x2_vals = [F_i(y,a,b,c) for y in y2_vals]
```

```python
plt.plot(x2_vals, y2_vals, color='blue')
plt.show()
```
