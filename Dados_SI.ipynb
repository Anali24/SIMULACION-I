```python
def volados(n):
  prob=[]
  l=[]
  tres=0
  for i in range(n):
    x = rnd.random()
    if 0.33<x<=0.49:
      tres +=1
      l.append(i)
      p= tres/(i+1) #probabilidad
      prob.append(p) # agragamos las probabiliades a la lista
  #grafica


  plt.plot(l,prob)
  plt.show()

  return prob
```
