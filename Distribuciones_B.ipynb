````python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
````

Parámetros

````python
a=0
b=1
c=2
````

Función de Densidad


Esta se obtiene mediante la gráfica, con ayuda de la recta tangente

````python
def f(x,a,b,c):
  if a<=x<b:
    return 5/4-x
  elif b<=x<=c:
    return 1/4
  else:
    return 0
````

Nueva forma de expresar el for:

```
y_vals = [fT(x) for x in x_vals]

```

es equivalente a:

```
y_vals = []
for x in x_vals:
    y_vals.append(fT(x))
```



````python
x_vals = np.linspace(0, 2, 1000000)
y_vals = [f(x,a,b,c) for x in x_vals]
````

````python
plt.plot(x_vals, y_vals, color='blue')
plt.show()
````

Funcion Acumulada

Se hace con la función de Densidad, y la derivamos


````python
def F(x,a,b,c):
  if a<=x<b:
    return (5/4)*x-(x**2)/2
  elif b<=x<=c:
    return (1/4)*x+1/2
  else:
    return 0
````

````python
x1_vals = np.linspace(0, 2, 1000000)
y1_vals = [F(x,a,b,c) for x in x1_vals]
````

````python
plt.plot(x1_vals, y1_vals, color='blue')
plt.show()
````

Función Inversa

Para la función inversa, igualamos a R y despejamos a x, al final R--->u


````python
def F_i(y):
  if 0<=y<1/4:
    return (5/4)+(25/16-2*y)**(1/2)
  elif 1/4<=y<=1:
    return 4*y-2
````

````python
y2_vals = np.linspace(0, 1, 1000000)
x2_vals = [F_i(y) for y in y2_vals]
````

````python
plt.plot(x2_vals, y2_vals, color='blue')
plt.show()
````
